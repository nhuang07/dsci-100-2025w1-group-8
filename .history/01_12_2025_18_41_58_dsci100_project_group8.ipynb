## Loading data into R and wrangling

**Reading in players and session data using link directly from GitHub**

```r
players_url <- "https://raw.githubusercontent.com/nhuang07/dsci-100-2025w1-group-8/refs/heads/main/players%20(1).csv"
players <- read_csv(players_url)
head(players)
```

**Tidying the data**

The players dataset is already tidy, so there will be no tidying work done for it. However, for our question, since we are using the "gender" and "experience" variables as predictors, we will need to convert their types to "factor." This work is seen below.

```r
players <- players |>
                mutate(experience = as_factor(experience)) |>
                mutate(gender = as_factor(gender))
head(players)
```
